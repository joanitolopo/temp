```python
import torch
import pandas as pd
from tqdm import tqdm

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
tokenizer = AutoTokenizer.from_pretrained("indonlp/cendol-mt5-small-chat", truncation_side='right', trust_remote_code=True)
tokenizer.padding_side = "right"
if tokenizer.pad_token is None:
  tokenizer.pad_token = tokenizer.bos_token if tokenizer.bos_token is not None else tokenizer.eos_token
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained("indonlp/cendol-mt5-small-chat", resume_download=True)
model.to(device)
```

```python
def predict_generation(prompts, tokenizer, model):
  inputs = tokenizer(prompts, return_tensors="pt", padding=True, truncation=True, max_length=1024).to(device)
  input_size = inputs["input_ids"].shape[1]

  outputs = model.generate(**inputs, min_length=1, max_length=100, no_repeat_ngram_size=2,
                           top_k=50, top_p=0.92, temperature=0.6, do_sample=True)
  preds = tokenizer.batch_decode(outputs, skip_special_tokens=True)

  return preds
```

```python
predict_generation("Hi! Nama kamu siapa?",tokenizer, model)
```
