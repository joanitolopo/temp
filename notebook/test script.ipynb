```python
from transformers import AutoTokenizer
from datasets import load_dataset
```

```python
tokenizer = AutoTokenizer.from_pretrained("indonlp/cendol-mt5-base-inst", padding_side="right", use_fast=True)
```

```python
data = load_dataset("joanitolopo/KupangMalay-InstructCorpus-v2", token="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")
```

```python
data['train']['prompt'][2]
```

```python
max_tokens = max(len(tokenizer.encode(text)) for text in data["train"]['prompt'])
print(f"Max tokens in dataset: {max_tokens}")
```

```python
import matplotlib.pyplot as plt

token_lengths = [len(tokenizer.encode(text)) for text in data["train"]['prompt']]
plt.hist(token_lengths, bins=20)
plt.title("Token Length Distribution")
plt.xlabel("Token Length")
plt.ylabel("Frequency")
plt.show()
```

```python
print(f"Max Token Length: {max(token_lengths)}")
print(f"Average Token Length: {sum(token_lengths)/len(token_lengths)}")
```
